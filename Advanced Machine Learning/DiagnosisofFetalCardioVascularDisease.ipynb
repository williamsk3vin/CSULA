```python
#installing necessary libraries
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from  sklearn.model_selection  import  GridSearchCV
from sklearn import preprocessing
```

```python
#Part A
df = pd.read_csv("https://github.com/mpourhoma/CS4662/raw/master/CTG_clean.csv")
df.head()
```

```python
features = ['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Min','Max',
           'Nmax','Nzeros','Mode','Mean','Median','Variance','Tendency']
X = df[features]
y = df['NSP']
```

```python
#Part B
scale_X = preprocessing.scale(X)
```

```python
#Part C
#Neural Network
kevins_ANN = MLPClassifier(hidden_layer_sizes = (30,), activation = 'logistic', solver = 'adam', alpha = 1,
                learning_rate_init = 0.02, random_state = 1, verbose = True)
```

```python
scores = cross_val_score(kevins_ANN, scale_X, y , cv = 10, scoring = 'accuracy')
```

```python
print(scores)
```

```python
means = scores.mean()
print(means)
```

```python
#Part D
seed = 1
np.random.seed(seed)
neuron_number = [(i,) for i in range(5,251,5)]

param_grid = dict(hidden_layer_sizes= neuron_number)
grid = GridSearchCV(kevins_ANN,param_grid=param_grid, cv=10,scoring = 'accuracy',verbose=True)
grid.fit(scale_X,y)
```

```python
print('The best params are:', grid.best_params_ )
```

```python
score = grid.best_estimator_.score(scale_X, y)
print('The accuracy is: ', score)
```
